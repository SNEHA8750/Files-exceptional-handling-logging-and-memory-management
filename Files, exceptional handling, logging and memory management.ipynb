```python
# Write a Python script that reads from one file and writes its content to another file.
# Python script to copy content from one file to another

# Specify file names
source_file = "source.txt"       # File to read from
destination_file = "destination.txt"  # File to write to

try:
    # Open source file in read mode and destination file in write mode
    with open(source_file, "r") as src, open(destination_file, "w") as dest:
        # Read content from source and write to destination
        content = src.read()
        dest.write(content)
    print("File copied successfully!")

except FileNotFoundError:
    print(f"Error: The file '{source_file}' does not exist.")
```

```python
# Write a Python program that logs an error message to a log file when a division by zero exception occurs.
import logging

# Configure logging
logging.basicConfig(
    filename="error_log.txt",      # Log file name
    level=logging.ERROR,           # Only log errors and above
    format="%(asctime)s - %(levelname)s - %(message)s"
)

try:
    numerator = int(input("Enter numerator: "))
    denominator = int(input("Enter denominator: "))

    result = numerator / denominator
    print("Result:", result)

except ZeroDivisionError:
    logging.error("Attempted division by zero!")  # Log the error
    print("Error: Division by zero is not allowed. Check error_log.txt for details.")
```

```python
# Write a program to handle a file opening error using exception handling.
# Program to handle file opening error

filename = "data.txt"  # File to open

try:
    with open(filename, "r") as file:
        content = file.read()
        print("File content:")
        print(content)

except FileNotFoundError:
    print(f"Error: The file '{filename}' was not found.")

except PermissionError:
    print(f"Error: You don't have permission to open '{filename}'.")

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# Write a program that uses the logging module to log both informational and error messages.
import logging

# Configure logging
logging.basicConfig(
    filename="app_log.txt",        # Log file name
    level=logging.INFO,            # Log INFO and above (INFO, WARNING, ERROR, CRITICAL)
    format="%(asctime)s - %(levelname)s - %(message)s"
)

try:
    logging.info("Program started.")  # Log informational message

    numerator = int(input("Enter numerator: "))
    denominator = int(input("Enter denominator: "))

    result = numerator / denominator
    print("Result:", result)
    logging.info("Division successful.")  # Log successful operation

except ZeroDivisionError:
    logging.error("Division by zero error occurred!")  # Log error message
    print("Error: Division by zero is not allowed.")

except ValueError:
    logging.error("Invalid input: Non-integer value entered!")  # Log error message
    print("Error: Please enter valid numbers.")

finally:
    logging.info("Program execution completed.")  # Log final info
```

```python
# How would you implement a basic logging setup that logs to a file with rotation after 1MB.
import logging
from logging.handlers import RotatingFileHandler

# Configure rotating log
log_file = "app_rotating.log"

handler = RotatingFileHandler(
    log_file,
    maxBytes=1*1024*1024,  # 1MB per log file
    backupCount=3          # Keep last 3 old log files
)

# Set logging format
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    handlers=[handler]
)

# Example logs
for i in range(10000):
    logging.info(f"Log message {i+1}")
```

```python
# How would you open a file and read its contents using a context manager in Python.
filename = "data.txt"

try:
    # Using a context manager to open and read the file
    with open(filename, "r") as file:
        content = file.read()  # Reads the entire file
        print("File content:")
        print(content)

except FileNotFoundError:
    print(f"Error: The file '{filename}' does not exist.")

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
#  Write a Python program that reads a file and prints the number of occurrences of a specific word.
filename = "data.txt"  # File to read
word_to_count = input("Enter the word to count: ").strip().lower()

try:
    with open(filename, "r") as file:
        content = file.read().lower()  # Convert to lowercase for case-insensitive match

        # Count occurrences of the word
        occurrences = content.split().count(word_to_count)

        print(f"The word '{word_to_count}' occurs {occurrences} times in the file.")

except FileNotFoundError:
    print(f"Error: The file '{filename}' does not exist.")

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# How can you check if a file is empty before attempting to read its contents.
import os

filename = "data.txt"

if os.path.exists(filename):
    if os.path.getsize(filename) == 0:
        print("The file is empty.")
    else:
        with open(filename, "r") as file:
            print(file.read())
else:
    print(f"Error: The file '{filename}' does not exist.")
```

```python
# Write a Python program that writes to a log file when an error occurs during file handling.
import logging

# Configure logging
logging.basicConfig(
    filename="file_errors.log",         # Log file name
    level=logging.ERROR,                # Only log errors
    format="%(asctime)s - %(levelname)s - %(message)s"
)

filename = "data.txt"

try:
    # Attempt to open and read the file
    with open(filename, "r") as file:
        content = file.read()
        print("File content:")
        print(content)

except FileNotFoundError:
    logging.error(f"File '{filename}' not found!")  # Log error to file
    print(f"Error: The file '{filename}' does not exist.")

except PermissionError:
    logging.error(f"No permission to access '{filename}'!")  # Log error
    print(f"Error: You do not have permission to access '{filename}'.")

except Exception as e:
    logging.error(f"Unexpected error with file '{filename}': {e}")  # Log any other error
    print(f"An unexpected error occurred: {e}")
```
